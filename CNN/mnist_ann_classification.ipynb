# Define the Model

```python


class FFN(nn.Module):
    def __init__(self,input_szs, output_szs, layer = [120,84]):
        super().__init__()
        #first layer
        self.fc1 = nn.ReLU(nn.Linear(input_szs,layer[0]))
        #second layer
        self.fc2 = nn.ReLU(nn.Linear(layer[0],layer[1]))
        #third layer
        self.fc3 = nn.Linear(layer[1],output_szs)
        
    
    def forward(self,x):
        x = self.fc1(x)
        x = self.fc2(x)
        x = self.fc3(x)
        return F.log_softmax(x, dim =1) # log_softmax coz we have multiple classes, softmax only when there are two classes
    
```

```python
# https://discuss.pytorch.org/t/whats-the-difference-between-nn-relu-vs-f-relu/27599/2

class FFN_(nn.Module):
    def __init__(self,input_szs, out_szs, layer=[120,84]):
        super().__init__()
        layerslist = []
        for i,szs in enumerate(layer):
            layerslist.append(nn.Linear(input_szs,szs))
            layerslist.append(nn.ReLU(inplace =True))
            input_szs = szs
        
        layerslist.append(nn.Linear(layer[-1],out_szs))
        print(layerslist)
        self.layers = nn.Sequential(*layerslist)
    
    def forward(self,x):
        x = self.layers(x)
        return F.log_softmax(x)
        
                
                
                
                
                
```
